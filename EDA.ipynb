# Univariate Analysis
univariate analysis focueses on analyzing each features in the dataset inedepandantly

. Distribution analysis : the distribution of each features is examined to identify its shape .control tendency and dispersion.

Identify potentioal issues : univariate anlysis is helps in identifying problems with the data such as outliers skewness ans missing values

THe shape of a data distribution referes to its overall patterns or forms as its represented on graph  . some comman shapes of data distribution includes:

.numerical distribution : a symmetrical and bell shaped distribution where the mean median and mode are equal and the majority of the data fails in the middle of the distribution with gradually decreasing frequencies towards the tails

| skewed distribution : A distribution that is not symetrical with one tail being longer than other it can either positively skewed or negetive skewed

| Bimodal distribution : A distribution with two peaks or modes.

| uniform distribution : A distribution where each values have an equal chance of occuring

The shape of the data distribution is important in identiying the presence of outliers skewness and the type of statistical tests and models that can be used for further analysis

| Dispersion is a statistical term used to spread or variability of a set data . measures how far the values in a data set are spread out form the central tendency (mean , median , mode) of the data.

there are several measures of dispresion including 

. Range : the diffrence b/w the largest and smallest values in a data set 

. variance : the avarage of the squared deviation of each values form of the data set

. standard deviation : the square root of the variance . it provides a measures of the spread of the data set is in the same unites as the original data 

. interquartiale range : the range between the first quartile and the thired quartile of the data

# steps of doing univariate Analysis on categorical columns

* visulization : create visulization to explore the distribution of the categoricas. some comman visulization for categorical data includes count plots and pie charts. this visulizations provides a visual representation of the distrubution of the categories and can helps identify any patterns or anomalies in the data
* missing values : check for missing values in the data decide how to handle them . Missing values can be imputed or excluded from the anlyis depending on the research question and the data set 

* conclusion : summarize the finding of the EDA and make decision about how to proceed with further analysis.
