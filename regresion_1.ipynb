```python
# label encoding for converting text data into a true flase 0/1 to use for over prediction and feature selection 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['frunishing_staues_0_1'] = le.fit_transform(df['furnishingstatus'])
df['mainroad'] = le.fit_transform(df['mainroad'])
df['airconditioning_0/1'] = le.fit_transform(df['airconditioning'])
#df['guestroom'] = le.fit_transform(df['guestroom'])
df['basment'] = le.fit_transform(df['basement'])
```

```python
X = df[['area',"bedrooms", "bathrooms", "stories", 'frunishing_staues_0_1', 'mainroad','parking','airconditioning_0/1']]
X 
```
